## python 객체로 변환하기

```python
# 네이버 검색 API 예제 - 블로그 검색
import json
import os
import sys
import urllib.request
client_id = "AoakXDTB44bQrc9XdCnw"
client_secret = "PpxHgwXhhS"
encText = urllib.parse.quote("송도맛집")
url = "https://openapi.naver.com/v1/search/blog?query=" + encText + "&display=100" # JSON 결과
# url = "https://openapi.naver.com/v1/search/blog.xml?query=" + encText # XML 결과
request = urllib.request.Request(url)
request.add_header("X-Naver-Client-Id",client_id)
request.add_header("X-Naver-Client-Secret",client_secret)
response = urllib.request.urlopen(request)
rescode = response.getcode()
if(rescode==200):
    response_body = response.read()
    result = json.loads(response_body.decode('utf-8'))
    print(result)
else:
    print("Error Code:" + rescode)
```

```python
result['items'][0]
```

```python
result['items'][0]['title']
```

```python
item_list = result['items']

title_list, link_list, date_list = [],[],[]

for i in range(len(item_list)):
    title = item_list[i]['title']
    link = item_list[i]['link']
    postdate = item_list[i]['postdate']
    
    title_list.append(title)
    link_list.append(link)
    date_list.append(postdate)

print(title_list[:3])
```

```python
import pandas as pd
item_result = pd.DataFrame({'제목':title_list, '링크': link_list,'날짜':date_list})
item_result.head()
```
