```python
import csv

f = open('incheon_population.csv','r')
data = csv.reader(f)
header = next(data)

result=[]
for row in data:
    if '송도1동' in row[0]:
        for male, female in zip(row[3:104],row[106:207]):
            result.append(int(male)+int(female))
f.close()

import matplotlib.pyplot as plt

plt.rc('font',family='Malgun Gothic')
plt.title('송도1동 연령별 인구수')
plt.bar(range(101),result)
plt.xlabel('연령')
plt.ylabel('인구수')
plt.show()
```

```python
import csv

f = open('incheon_population.csv','r')
data = csv.reader(f)
header = next(data)

male = []
female = []

for row in data:
    if '송도1동' in row[0]:
        for i in row[3:104]:
            male.append(-int(i))
        for i in row[106:207]:
            female.append(int(i))
f.close()

import matplotlib.pyplot as plt

plt.rc('font',family='Malgun Gothic')
plt.style.use('ggplot')
plt.title('송도1동 연령별 인구수')
plt.rcParams['axes.unicode_minus'] = False
plt.barh(range(101),male,label='남성')
plt.barh(range(101),female,label='여성')
plt.legend()
plt.xlabel('연령')
plt.ylabel('인구수')
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt

f = open('incheon_population.csv','r')
data = csv.reader(f)
header = next(data)

male =[]
female =[]

name = input('연령별 인구수를 알고싶은 동네: ')
for row in data:
    if name in row[0]:
        for i in row[3:104]:
            male.append(-int(i))
        for i in row[106:207]:
            female.append(int(i))
f.close()

plt.rc('font',family='Malgun Gothic')
plt.title(name+' 연령별 인구수')
plt.style.use('ggplot')
plt.rcParams['axes.unicode_minus'] = False
plt.barh(range(101),male,label='남자')
plt.barh(range(101),female,label='여자')
plt.xlabel('연령')
plt.ylabel('인구수')
plt.legend()
plt.show()
```
