```python
import matplotlib.pyplot as plt

plt.style.use('ggplot')
x = [1,2,3,4]
y = [10,30,20,40]
size = [100,200,300,400]
color = ['r','b','g','gold']
plt.scatter(x,y,s=size,c=color)
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.style.use('ggplot')
x = [1,2,3,4]
y = [10,30,20,40]
size = [100,200,300,400]
plt.scatter(x,y,s=size,c=range(4))
plt.colorbar()
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.style.use('ggplot')
x = [1,2,3,4]
y = [10,30,20,40]
size = [100,200,300,400]
plt.scatter(x,y,s=size,c=range(4),cmap='jet')
plt.colorbar()
plt.show()
```

```python
import matplotlib.pyplot as plt
import random
x,y = [],[]
size=[]
for i in range(100):
    x.append(random.randint(50,100))
    y.append(random.randint(50,100))
    size.append(random.randint(10,100))
    
plt.scatter(x,y,s=size,c=size,cmap='jet')
plt.show()
```

```python
import matplotlib.pyplot as plt
import random
x,y = [],[]
size=[]
for i in range(100):
    x.append(random.randint(50,100))
    y.append(random.randint(50,100))
    size.append(random.randint(10,100))
    
plt.scatter(x,y,s=size,c=size,cmap='jet',alpha=0.3)
plt.show()
```
