```python
import pandas as pd
url = "https://ko.wikipedia.org/wiki/%EC%98%AC%EB%A6%BC%ED%94%BD_%EB%A9%94%EB%8B%AC_%EC%A7%91%EA%B3%84"
df = pd.read_html(url)
df
```

```python
df = pd.read_html(url, header=0,index_col=0) #헤더를 열이름으로, 나라이름을 행이름으로 지정하기위해 설정
df[0]
```

```python
# 슬라이싱을 활욜한 동계올림픽에 대한 데이터 추출
a= df[0].iloc[:,5:10] #행은 전체 열은 5~9까지 (a= df[0].iloc[0:len(),0:5])
a
```

```python
# 컬럼명 중 unnamed 부분 컬럼명 설정하기
a.columns = ['동계 참가 횟수','금','은','동','계']
a
```

```python
# 참가횟수, 금메달 기준 내림차순 정렬하기
a= a.sort_values(by=['동계 참가 횟수','금'],ascending=False)
a
```

```python
#csv 파일로 저장하기
a.to_csv('winter_medal.csv')
```
