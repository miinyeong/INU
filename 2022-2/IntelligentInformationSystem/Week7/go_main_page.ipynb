```python
from bs4 import BeautifulSoup
from urllib.request import urlopen

url = "https://www.naver.com"
html = urlopen(url)

soup = BeautifulSoup(html.read(),"html.parser")  

result = soup.find("div",{"class":"service_area"})
a_result = result.find("a")
print(a_result)
print(a_result.text) ##    
```
