```python
from bs4 import BeautifulSoup
from urllib.request import urlopen

url = "https://movie.naver.com/movie/sdb/rank/rmovie.naver?sel=pnt&date=20221018"
html = urlopen(url)

soup = BeautifulSoup(html.read(),"html.parser")
title_list = soup.find_all("div",{"class":"tit5"})

for title in title_list:
    print(title.text)                ## 태그를 떼어냈을 때 공백으로 바뀐 부분도
                                    ##포함하기때문에 공백도 출력
```

```python
from bs4 import BeautifulSoup
from urllib.request import urlopen

url = "https://movie.naver.com/movie/sdb/rank/rmovie.naver?sel=pnt&date=20221018"
html = urlopen(url)

soup = BeautifulSoup(html.read(),"html.parser")
title_list = soup.find_all("div",{"class":"tit5"})

for title in title_list:
    print(title.text.strip("\n"))      ## 문자열 앞뒤로 연속된 모든 \n을 삭제
```

```python
from bs4 import BeautifulSoup
from urllib.request import urlopen

url = "https://movie.naver.com/movie/sdb/rank/rmovie.naver?sel=pnt&date=20221018"
html = urlopen(url)

soup = BeautifulSoup(html.read(),"html.parser")
title_list = soup.find_all("div",{"class":"tit5"})

for title in title_list:
    print(title.get_text(strip="\n"))   ## print(title.text.strip("\n"))와 동일
```

# 평점 값 가져오기

```python
from bs4 import BeautifulSoup
from urllib.request import urlopen

url = "https://movie.naver.com/movie/sdb/rank/rmovie.naver?sel=pnt&date=20221018"
html = urlopen(url)

soup = BeautifulSoup(html.read(),"html.parser")
point_list = soup.find_all("td",{"class":"point"})
print(point_list)
```

```python
from bs4 import BeautifulSoup
from urllib.request import urlopen

url = "https://movie.naver.com/movie/sdb/rank/rmovie.naver?sel=pnt&date=20221018"
html = urlopen(url)

soup = BeautifulSoup(html.read(),"html.parser")
point_list = soup.find_all("td",{"class":"point"})
for point in point_list:
    print(float(point.text))     ## 데이터 실수형 변환
```
