## 최고기온을 리스트에 저장

```python
import matplotlib.pyplot as plt

plt.plot(high_temp,color='r')
plt.show()
```

## 9월 13일의 최고 기온 데이터 시각화하기

```python
import csv

f=open('incheon_weather.csv','r')
data = csv.reader(f)
header = next(data)

high_temp = []
for row in data:
    if row[-1] != '':
        if(row[0].split('-')[1]=='09' and row[0].split('-')[2] =='13'):
            row[-1] = float(row[-1])
            high_temp.append(row[-1])
f.close()
print(high_temp)
```

## 1999년 기준으로 하여 최고기온 및 최저기온 데이터 구하기

```python
import csv

f=open('incheon_weather.csv','r')
data = csv.reader(f)
header = next(data)

high_temp = []
low_temp = []
for row in data:
    if row[0].split('-')[0] == '1999':
        if row[-1] != '':
            row[-1] = float(row[-1])
            high_temp.append(row[-1])
        if row[-2] != '':
            row[-2] = float(row[-2])
            low_temp.append(row[-2])
f.close()
```

```python
import matplotlib.pyplot as plt
plt.rc('font',family = 'Malgun Gothic')
plt.plot(high_temp,color='pink', label='최고기온',ls=":")
plt.plot(low_temp,color='skyblue', label='최저기온',ls="--")
plt.legend()
plt.show()
print(len(high_temp))
```

## 1999년 9월 13일 부터 현재까지 매년 9월 13일의 데이터 추출 및 시각화

```python
import csv

f=open('incheon_weather.csv','r')
data = csv.reader(f)
header = next(data)

high_temp = []
low_temp = []
for row in data:
    if row[0].split('-')[0] >= '1999':
        if row[0].split('-')[1] == '09' and row[0].split('-')[2] == '13':
            if row[-1] != '' and row[-2] != '':
                row[-1] = float(row[-1])
                high_temp.append(row[-1])
                row[-2] = float(row[-2])
                low_temp.append(row[-2])
f.close()
```

```python
import matplotlib.pyplot as plt
plt.rcParams['axes.unicode_minus'] = False
plt.rc('font',family = 'Malgun Gothic')
plt.title('1999년 이후 9월 13일의 최고,최저기온')
x = list(row(0))
plt.plot(x,high_temp,color='pink',label='최고기온')
plt.plot(x,low_temp,color='skyblue',label='최저기온')
plt.legend(loc=1)
plt.xticks(x, rotation=60)
plt.legend()
```

# 9월 13일의 최고 기온 데이터 시각화하기

```python
import csv
f = open('incheon_weather.csv','r')
data = csv.reader(f)
header = next(data)
max_temp = []
for row in data:
    if row[-1] != '':
        if row[0].split('-')[1] == '09' and row[0].split('-')[2] == '13':
            max_temp.append(float(row[-1]))
f.close()
print(max_temp)
```

```python
import matplotlib.pyplot as plt
plt.rc('font', family='Malgun Gothic')
plt.title('9월 13일 최고 기온 데이터 시각화')
plt.plot(max_temp,color='hotpink')
plt.show()
```

```python
import csv
f = open('incheon_weather.csv','r')
data = csv.reader(f)
header = next(data)
max_temp = []
min_temp = []
for row in data:
    if row[-1] != '':
        if row[0].split('-')[0] >= '1999':
            if row[0].split('-')[1] == '02' and row[0].split('-')[2] == '13':
                max_temp.append(float(row[-1]))
                min_temp.append(float(row[-2]))
f.close()
print(max_temp,min_temp)
```

```python
import matplotlib.pyplot as plt
plt.rcParams['axes.unicode_minus'] = False
plt.rc('font', family='Malgun Gothic')
plt.title('9월 13일 최고,최저 기온 데이터 시각화')
plt.plot(max_temp,color='hotpink')
plt.plot(min_temp,color='skyblue')
plt.show()
```
