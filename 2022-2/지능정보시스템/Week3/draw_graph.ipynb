## 범례 : 그래프가 의미하는 바를 구별하기 위해 사용

```python
import matplotlib.pyplot as plt

plt.rc('font',family = 'Malgun Gothic')
plt.title('플러팅')
plt.plot([10,20,30,40],label='범례1')
plt.plot([55,22,66,20],label='범례1')
plt.plot([34,54,72,20],label='범례1')
plt.legend(loc=5)
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.rc('font',family = 'Malgun Gothic')
plt.title('플러팅')
plt.plot([10,20,30,40],label='범례1',color='skyblue')
plt.plot([34,55,76,22],label='범례2', color='hotpink')
plt.legend(loc=5)
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.rc('font',family = 'Malgun Gothic')
plt.title('플러팅')
plt.plot([10,20,30,40],label='범례1',color='skyblue',ls=':')
plt.plot([34,55,76,22],label='범례2', color='hotpink',ls="--")
plt.legend(loc=5)
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.rc('font',family = 'Malgun Gothic')
plt.title('플러팅')
plt.plot([10,20,30,40],label='범례1',color='skyblue',ls=':',marker='v')
plt.plot([34,55,76,22],label='범례2', color='hotpink',ls="--",marker='o')
plt.legend(loc=5)
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.rc('font', family='Malgun Gothic')
plt.title('플러팅')
plt.plot([10,20,30,40],label='범례1',ls=':',color='skyblue',marker='v')
plt.plot([34,55,76,22],label='범례2',ls='--',color='pink',marker='o')
plt.legend(loc=5)
plt.show()
```
