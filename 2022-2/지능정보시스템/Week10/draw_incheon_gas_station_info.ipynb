```python
from glob import glob
```

## 파일명 리스트 station_files라는 변수에 저장

```python
station_files = glob('지역*.xls')
station_files
```

## 엑셀 파일 불러오기

```python
import pandas as pd

tmp_list = []

for file_name in station_files:
    tmp = pd.read_excel(file_name, header=2)
    tmp_list.append(tmp)
    
tmp_list[0]
```

```python
stations = pd.concat(tmp_list)
stations.shape
```

## 원하는 컬럼만 가져오기

```python
stations_sel = stations.loc[:,['상호','주소','상표','셀프여부','휘발유','경유']]
stations_sel
```

## 컬럼명 변경하기

```python
stations_sel.columns = ['주유소','주소','상표','셀프여부','휘발유','경유']
stations_sel
```

## 군/구 컬럼 추가

```python
gu_list  = []

for address in stations['주소']:
    address_gu = address.split()[1]
    gu_list.append(address_gu)
    
stations_sel['군/구'] = gu_list
stations_sel
```

```python
stations_sel1 = stations_sel[['상표','주유소','주소','셀프여부','휘발유','군/구']]
stations_sel1
```

## 종류 컬럼 추가하기

```python
type_list1  = '휘발유'

stations_sel1['종류'] = type_list1
stations_sel1  
```

```python
stations_sel1 = stations_sel1[['상표','주유소','주소','셀프여부','휘발유','종류','군/구']]
stations_sel1
```

```python
stations_sel1.columns = ['상표','주유소','주소','셀프여부','가격','종류','군/구']
stations_sel1
```

```python
stations_sel2 = stations_sel[['상표','주유소','주소','셀프여부','경유','군/구']]
stations_sel2
```

```python
type_list2  = '경유'

stations_sel2['종류'] = type_list2
stations_sel2  
```

```python
stations_sel2 = stations_sel2[['상표','주유소','주소','셀프여부','경유','종류','군/구']]
stations_sel2
```

```python
stations_sel2.columns = ['상표','주유소','주소','셀프여부','가격','종류','군/구']
stations_sel2
```

## 데이터프레임 합치기

```python
import pandas as pd
```

```python
stations_sel = pd.concat([stations_sel1,stations_sel2])
stations_sel
```

## 가격이 기입되어 있지 않은 주유소 제외하기

```python
stations_sel = stations_sel.query("가격 != '-'")
stations_sel
```

## 가격을 숫자(int)로 변경하기

```python
price_list = []

for price in stations_sel['가격']:
    price_new = int(price)
    price_list.append(price_new)
    
stations_sel['가격'] = price_list
stations_sel
```

```python
stations_sel = stations_sel.query("가격 != '-'").copy() # 경고 해결하기
stations_sel
```

```python
price_list = []

for price in stations_sel['가격']:
    price_new = int(price)
    price_list.append(price_new)
    
stations_sel['가격'] = price_list
stations_sel
```

```python
stations_sel.reset_index(inplace=True)
stations_sel
```

## 그래프1) x축: 종류, y축: 가격, 색상: 셀프여부

```python
import matplotlib.pyplot as plt

plt.rc('font',family = 'Malgun Gothic')
plt.rcParams['axes.unicode_minus'] = False
plt.style.use("ggplot")

import seaborn as sns

plt.figure(figsize=(12,8))
sns.boxplot(x='종류',y="가격",hue='셀프여부',data = stations_sel,palette="Set3")
plt.show()
```

### 셀프여부에 상관없이 휘발유의 가격이 경유의 가격보다 낮게 나왔다.
### 휘발유에서 셀프여부가 Y인 경우의 가격의 중앙값은 1500후반이고 최소값은 약 1550이며 최대값은 약 1650이다. 셀프여부 N인 경우의 가격의 중앙값은 1600과 1650 사이이며 최소값은 약 1550, 최대값은 약 1800이다.
### 경유에서 셀프여부가 Y인 경우의 가격의 중앙값은 1850이고 최소값은 1800이며 최대값은 약 1950이다. 셀프여부 N인 경우의 가격의 중앙값은 1800후반이며 최소값은 약 1800, 최대값은 2000초과이다.
### 휘발유과 경유 모두 셀프여부 Y에서의 1사분위수,3사분위수가 셀프여부 N의 사분위수보다 낮다.

## 그래프2) x축: 상표, y축: 가격, 색상: 종류

```python
plt.figure(figsize=(12,8))
sns.boxplot(x='상표',y="가격",hue='종류',data = stations_sel,palette="summer")
plt.show()
```

### 모든 주유소에서 휘발유보다 경유의 가격이 높게 나왔고, 휘발유의 경우 최대값과 최소값의 차이가 가장 큰 주유소는 자가상표이며, 
### 경유에서는 현대오일뱅크가 가격이 가장 크다.
### 휘발유와 경우 둘다 가격의 중앙값이 가장 작은 주유소는 알뜰 주유소, 가장 큰 주유소는 자가상표이다. 각 주유소의 1사분위수, 3사분위수 역시 가장 낮은 주유소는 알뜰 주유소, 가장 높은 주유소는 자가상표이다.
### 다른 주유소에 비해 자가상표에서 휘발유와 경유의 가격이 상대적으로 크게 나타난다.

## 그래프2) x축: 군/구, y축: 가격, 색상: 종류

```python
plt.figure(figsize=(12,8))
sns.boxplot(x='군/구',y="가격",hue='종류',data = stations_sel,palette="summer")
plt.show()
```

### 모든 주유소에서 휘발유보다 경유의 가격이 높게 나왔고, 휘발유의 경우 최대값과 최소값의 차이가 가장 큰 군/구는 중구이며,
### 경유에서도 중구가 가장 차이가 크다.
### 대부분 군/구에서 휘발유 또는 경유의 차이가 크게 나진 않지만 옹진군에서의 휘발유와 경유가 타지역에 비해 상대적으로 가격이 크게 나타난다.
### 휘발유, 경유 모두 각  군/구에서의 사분위수에서 동구의 1,3사분위가 가장 낮고, 옹진군의 1,3사분위수가 가장 높다.
