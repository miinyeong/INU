```python
import matplotlib.pyplot as plt
plt.rc('font',family='Malgun Gothic')
plt.axis('equal')
size = [120,250,330,410]
label=['레이블1','레이블2','레이블3','레이블4']
plt.pie(size,labels=label, autopct='%.1f%%')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.rc('font',family='Malgun Gothic')
plt.axis('equal') # 표를 원형으로 만들기
size = [120,250,330,410] #각 레이블 크기
label=['레이블1','레이블2','레이블3','레이블4'] # 레이블 이름 지정
color = ['darkmagenta','deeppink','hotpink','pink'] # 각 레이블 색 지정
exp = [0,0,0.1,0] # 돌출효과
plt.pie(size,labels=label, autopct='%.1f%%',colors=color,explode=exp)
plt.legend()
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt

f=open('incheon_population.csv','r')
data = csv.reader(f)

result=[]
m_sum=0
f_sum=0


for row in data:
    if '송도1동' in row[0]:
        for i in range(101):
            m_sum += int(row[i+3])
            f_sum += int(row[i+106])
            
result.append(m_sum)
result.append(f_sum)
f.close()

plt.rc('font',family='Malgun Gothic')
plt.title('송도1동 성별 인구분포')
plt.axis('equal')

label=['남','여']
color = ['crimson','darkcyan']
plt.pie(result, colors=color,labels=label,autopct='%.1f%%',startangle=90)
plt.legend()
plt.show()
```
