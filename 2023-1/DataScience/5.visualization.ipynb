### Matplotlib
#### Plotting
The most important function in `matplotlib` is plot, which allows you to plot 2D data.

```python
# Compute the x and y coordinates for points on a sine curve
x = np.arange(0, 3 * np.pi, 0.3)
y = np.sin(x)
print(x)
print(y)
print(x.shape, y.shape)
```

```python
# Plot the points using matplotlib
plt.plot(x, y, 'D-.', color='c')

# Refefence
# https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html#
```

```python
# Plot the points using matplotlib
plt.plot(x, y, 'o')
```

```python
y_sin = np.sin(x)
y_cos = np.cos(x)

# Plot the points using matplotlib
plt.plot(x, y_sin, c='b')
plt.plot(x, y_cos, c='k')
plt.xlabel('x axis label', fontdict= {'family': 'serif',
                                      'color':  'darkred',
                                      'weight': 'normal',
                                      'size': 30,
                                      })
plt.ylabel('y axis label')
plt.title('Sine and Cosine')
plt.legend(['Sine', 'Cosine'], loc='lower right')
```

#### Subplots 
You can plot different things in the same figure using the subplot function.

```python
# Compute the x and y coordinates for points on sine and cosine curves
x = np.arange(0, 3 * np.pi, 0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)

# Set up a subplot grid that has height 2 and width 1,
# and set the first such subplot as active, and make the first plot.
plt.subplot(1, 2, 1)
plt.plot(x, y_sin)
plt.title('Sine')

# Set the second subplot as active, and make the second plot.
plt.subplot(1, 2, 2)
plt.plot(x, y_cos)
plt.title('Cosine')

# Show the figure.
plt.show()
```

#### Data visualization with Seaborn

```python
# barplot
plt.figure(figsize=(10,8))
sns.lmplot(x='Attack', y='Defense', 
           data=df,
           hue='Type 1')
plt.ylim(0, None)

# Reference
# https://seaborn.pydata.org/generated/seaborn.lmplot.html
```

```python
# boxplot
plt.figure(figsize=(10,8))

sns.boxplot(data=df)

# Reference
# https://seaborn.pydata.org/generated/seaborn.boxplot.html
```

```python
# drop Total, Generation and Legendary columns
plt.figure(figsize=(10, 8))

stats_df = df.drop(['Total', 'Generation','Legendary'], axis=1)
stats_df.head()
sns.boxplot(data=stats_df)
```

```python
# violin plot
# sns.set_style('white')
plt.figure(figsize=(12, 8))
sns.violinplot(x='Type 1', y='Attack', data=df)
# sns.violinplot(x='Type 1', y='Defense', data=df)

# Reference
# https://seaborn.pydata.org/generated/seaborn.boxplot.html
```

```python
# Correlation
corr = stats_df.corr()

# heatmap
sns.heatmap(corr)

# Reference
# https://seaborn.pydata.org/generated/seaborn.heatmap.html
```

```python
# histogram
sns.distplot(df.Attack, bins=10)

# Reference
# https://seaborn.pydata.org/generated/seaborn.distplot.html
```

```python
# countplot
sns.countplot(x='Type 1', data=df)
 
# Rotate x-labels
plt.xticks(rotation=-45)

# Reference
# https://seaborn.pydata.org/generated/seaborn.countplot.html
```
